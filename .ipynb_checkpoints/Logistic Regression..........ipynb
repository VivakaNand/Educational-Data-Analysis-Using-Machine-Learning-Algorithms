```python
import numpy as np
import pandas as pd
from numpy import genfromtxt
from sklearn import linear_model
file=genfromtxt("vivek.csv", delimiter=",", dtype="str")
#print(file)
dic={}
count=0
for val in file:
    if val[0] not in dic:
        dic[val[0]]= count
        count+=1
for val in file:
    val[0]=dic[val[0]]
    
count=0
for val in file:
    if val[1] not in dic:
        dic[val[1]]= count
        count+=1
for val in file:
    val[1]=dic[val[1]]
    
count=0
for val in file:
    if val[4] not in dic:
        dic[val[4]]= count
        count+=1
for val in file:
    val[4]=dic[val[4]]

count=0
for val in file:
    if val[5] not in dic:
        dic[val[5]]= count
        count+=1
for val in file:
    val[5]=dic[val[5]]

count=0
for val in file:
    if val[6] not in dic:
        dic[val[6]]= count
        count+=1
for val in file:
    val[6]=dic[val[6]]
   
    
trainingSet=file[:343]
testingSet=file[343:]

trainingX=trainingSet[:,[0,1,2,3,4,5,6,7,8]]
trainingX=trainingX.astype(float)
trainingY=trainingSet[:,[9]]

testingX=testingSet[:,[0,1,2,3,4,5,6,7,8]]
testingX=testingX.astype(float)
testingY=testingSet[:,[9]]
lr=linear_model.LogisticRegression()
lr.fit(trainingX, trainingY)


pred=lr.predict([[0,0,65,65,0,0,0,30.5,3]])
print("Predicted Value is"+ str(pred))

print('\n',  dic)
lr.score(testingX, testingY)*100
```
